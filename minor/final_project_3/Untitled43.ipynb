```python
from sklearn.preprocessing import StandardScaler

sc_x=StandardScaler()
sc_y=StandardScaler()
X_std=sc_x.fit_transform(X)
y_std=sc_y.fit_transform(y.reshape(-1,1)).flatten()
```

```python
plt.plot(y_std)
```

```python
from sklearn.model_selection import  train_test_split
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score, median_absolute_error
```

```python
X_train_scl,X_test_scl,y_train_scl,y_test_scl = train_test_split(X_std, y_std, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import Lasso
#L =(y-yO)**2+a*sum|w|-->min
mlasso=Lasso(alpha=0.1)
mlasso.fit(X_train_scl, y_train_scl)
```

```python
y_train_pr=mlasso.predict(X_train_scl)
```

```python
y_test_pr=mlasso.predict(X_test_scl)
```

```python
print(mlasso.coef_)
```

```python
print(mean_squared_error(y_train_scl,y_train_pr))
print(mean_squared_error(y_test_scl,y_test_pr))
```

```python
print(r2_score(y_train_scl,y_train_pr))
print(r2_score(y_test_scl,y_test_pr))
```

# Ridge

```python
from sklearn.linear_model import Ridge
#L =(y-yO)**2+a*sum(w**2)-->min
mridge=Ridge(alpha=0.1)
mridge.fit(X_train_scl, y_train_scl)
y_train_pr1=mridge.predict(X_train_scl)
y_test_pr1=mridge.predict(X_test_scl)
print(mean_squared_error(y_train_scl,y_train_pr1))
print(mean_squared_error(y_test_scl,y_test_pr1))
print(r2_score(y_train_scl,y_train_pr1))
print(r2_score(y_test_scl,y_test_pr1))
print(mridge.coef_)
```

# Elastic

```python
from sklearn.linear_model import ElasticNet
mel=ElasticNet(alpha=0.1, l1_ratio=0.5)
mel.fit(X_train_scl, y_train_scl)
y_train_pr2=mel.predict(X_train_scl)
y_test_pr2=mel.predict(X_test_scl)
print(mean_squared_error(y_train_scl,y_train_pr2))
print(mean_squared_error(y_test_scl,y_test_pr2))
print(r2_score(y_train_scl,y_train_pr2))
print(r2_score(y_test_scl,y_test_pr2))
print(mel.coef_)
```
